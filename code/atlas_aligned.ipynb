```python
import nibabel as nib
from nilearn import image, plotting
from matplotlib import pyplot as plt00

%matplotlib inline
```

```python
mni = nib.load('../data/MNI152_T1_1mm_brain.nii.gz')
mni.affine
```

```python
dkt = nib.load('./DKT_w_hipp_labels.nii.gz')
dkt2 = image.resample_img(dkt, mni.affine, mni.shape)
dkt.affine, dkt2.affine
```

```python
cho = nib.load('./Cho_Braak_w_hipp_labels.nii.gz')
cho.affine
cho2 = image.resample_img(cho, mni.affine, mni.shape)
cho.affine, cho2.affine
```

```python
plotting.plot_roi(dkt2, mni)
plotting.plot_roi(cho2, mni)
```

```python
nib.save(dkt2, './DKT_w_hipp_labels_gk.nii.gz')
nib.save(cho2, './Cho_Braak_w_hipp_labels_gk.nii.gz')
```
