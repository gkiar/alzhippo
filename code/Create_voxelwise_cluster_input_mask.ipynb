```python
import numpy as np
import nibabel as ni
import matplotlib.pyplot as plt
from nilearn import plotting, image
```

```python
atlas_img = ni.load('../data/dilated_hipp_parcellation_gspace.nii.gz')
atlas = atlas_img.get_data()
```

```python
plt.close()
plotting.plot_roi(atlas_img)
plt.show()
```

```python
n_vox = len(atlas[atlas>0])
clus_input = np.zeros_like(atlas)
clus_input[atlas>0] = range(1,n_vox+1)
```

```python
plt.close()
plotting.plot_roi(ni.Nifti1Image(clus_input,atlas_img.affine))
plt.show()
```

```python
print(len(np.unique(clus_input)),clus_input.max(), n_vox)
```

```python
clus_input.max() == n_vox
```

```python
ni.Nifti1Image(clus_input,atlas_img.affine).to_filename('../data/hippo_erc_voxelwise')
```
